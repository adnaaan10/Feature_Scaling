# **Standardization**

```python
df=df.iloc[:,2:]
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.1,random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
scalar.fit(X_train)
X_train_scalar=scalar.transform(X_train)
X_test_scalar=scalar.transform(X_test)
```

```python
X_train
```

```python
X_train_scalar
```

```python
X_test_scalar=pd.DataFrame(X_test_scalar,columns=X_train.columns)
X_train_scalar=pd.DataFrame(X_train_scalar,columns=X_test.columns)
```

```python
X_test_scalar
```

```python
X_train_scalar
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scalar['Age'],X_train_scalar['EstimatedSalary'])
ax2.set_title('After Scaling')
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(ncols=4,figsize=(12,6))
# Pass the DataFrame and specify 'x' and 'y' columns for kdeplot
sns.kdeplot(x=X_train['Age'], ax=ax1) # Specify ax=ax1 to plot on the first subplot
ax1.set_title(' Age Before Scaling')
sns.kdeplot(x=X_train_scalar['Age'], ax=ax2) # Specify ax=ax2 to plot on the second subplot
ax2.set_title('Age After Scaling')
sns.kdeplot(x=X_train['EstimatedSalary'], ax=ax3) # Specify ax=ax3 to plot on the third subplot
ax3.set_title('EstimatedSalary Before Scaling')
sns.kdeplot(x=X_train_scalar['EstimatedSalary'], ax=ax4) # Specify ax=ax4 to plot on the fourth subplot
ax4.set_title('EstimatedSalary After Scaling')
```

# **Normalization**

```python
df=pd.read_csv('/content/wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label','Alcohol','Malic acid']
```

```python
df
```

```python
sns.kdeplot(x=df['Alcohol'])
```

```python
sns.kdeplot(x=df['Malic acid'])
```

```python
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class label'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.1,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax=MinMaxScaler()
```

```python
minmax.fit(X_train)
```

```python
X_train_scaled=minmax.transform(X_train)
X_test_scaled=minmax.transform(X_test)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.scatter(X_train['Alcohol'],X_train['Malic acid'])
ax1.set_title('Before Scaling')
# Access the columns of the scaled data using numerical indices
ax2.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1])  # Assuming 'Alcohol' is column 0 and 'Malic acid' is column 1
ax2.set_title('After Scaling')
plt.show() # This line was missing in your original code, adding for display
```

```python
#we can convert numpy array to dataframe
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.scatter(X_train['Alcohol'],X_train['Malic acid'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic acid'])
ax2.set_title('After Scaling')
```
